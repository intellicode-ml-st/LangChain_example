# Memory (short term memory)

```python
from langgraph.checkpoint.memory import InMemorySaver
from langchain.chat_models import init_chat_model
from langchain.agents import create_agent
from pprint import pprint

tools = []

model = init_chat_model(
    model="llama3.2:3b", 
    model_provider="ollama",
    temperature=0.8,
)

agent = create_agent(
    model,
    tools,
    checkpointer=InMemorySaver()
)
```

```python
response = agent.invoke(
    {"messages": [{"role": "user", "content": "Hi, I'm Intellicode jtoh."}]},
    {"configurable": {"thread_id": "thread_1"}}
)

response["messages"][-1].content
```

```python
response = agent.invoke(
    {"messages": [{"role": "user", "content": "what is my name?"}]},
    {"configurable": {"thread_id": "thread_1"}} # 동일 thread_id 재사용
)

response["messages"][-1].content  # 이전 대화내용 기억 하는 것을 확인
```

```python
response = agent.invoke(
    {"messages": [{"role": "user", "content": "what is my name?"}]},
    {"configurable": {"thread_id": "thread_2"}} # 다른 thread_id 사용시 ?
)

response["messages"][-1].content # 모르는 내용이라 모델 자체의 답변이 옴
```

-- thread_id == 개인 카톡방 으로 생각하면 편리
-- 동일한 thread_id 를 가진 경우 대화 내용을 지속적으로 기억(memory)하게 됨

```python
response = agent.invoke(
    {"messages": [{"role": "user", "content": "what talking about? untill now."}]},
    {"configurable": {"thread_id": "thread_1"}}
)
```

```python
for i, msg in enumerate(response["messages"]):
    print(f"--- Message {i+1} ---")
    print(msg.content)
    print()
```
